# Köhler Curves

<br>
We want to plot the Köhler curves (Fig. 3.12). For this, we use equation 3.73 and assume constant values for $A_K$ and $B_K$ (for a temperature of 20°C):

```r
saturation_ratio = function(rd,iNs) {
    #@param: rd drop radius [m]
    #@param: iNs effective solute content [N_s in mol]
    #@return: saturation ratio
    Ak = 1.1*10^-9
    Bk = 4.3*10^-6
    return(Ak/rd - Bk*iNs/rd^3+1)
}

#plotting
rds=10^seq(-3,2,0.01)*10^(-6) #m
solution=10^seq(-20,-16) #mol
plot(rds,saturation_ratio(rds,solution[1]),ylim=c(0.95,1.07),type="l",log="x",ylab="Saturation ratio",
     xlab="Drop radius [m]",main="Köhler curves",lwd=2)
for (i in 2:length(solution)) { points(rds,saturation_ratio(rds,solution[i]),col=i,type="l",lwd=2) }
grid()
legend("topright",legend=solution,title="effective solute content [mol]:",col=c(1:length(solution)),
       lwd=2,lty=1,bg="white")
text(0.2e-8,1,"Raoult effect >\nKöhler effect",cex=0.9)
text(1e-6,1.02,"Köhler effect >\nRaoult effect",cex=0.9)
```
